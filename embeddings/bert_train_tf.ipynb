```python
import os
```

```python
import tokenizers
```

```python
bwpt = tokenizers.BertWordPieceTokenizer(
#     from_file = None,
#     add_special_tokens=True,
    unk_token='[UNK]',
    sep_token='[SEP]',
    cls_token='[CLS]',
    clean_text=True,
    handle_chinese_chars=True,
    strip_accents=True,
    lowercase=True,
    wordpieces_prefix='##'
)
```

```python
bwpt.train(
    files=["../../data_text/hi_dedup_1000.txt"],
    vocab_size=30000,
    min_frequency=3,
    limit_alphabet=1000,
    special_tokens=['[PAD]', '[UNK]', '[CLS]', '[MASK]', '[SEP]']
)
```

```python
bwpt_dict = {k: v for k, v in sorted(bwpt.get_vocab().items(), key=lambda item: item[1])}
```

```python
with open("../../data_text/hindi-vocab.txt", 'w') as file:
        for row in list(bwpt_dict.keys()):
            s = "".join(map(str, row))
            file.write(s+'\n')
```
